```python
%matplotlib inline
%config InlineBackend.figure_formats = ['svg']
#@markdown 例:連続で滑らかな関数 https://en.wikipedia.org/wiki/Lennard-Jones_potential
# ref https://towardsdatascience.com/the-lennard-jones-potential-35b2bae9446c
# ref https://qiita.com/code0327/items/d2f199c5e1267d4822fb
import numpy as np
import matplotlib.pyplot as plt

def ljp(r, epsilon, sigma):
    
    return 48 * epsilon * np.power(sigma, 12) / np.power(r, 13) \
    - 24 * epsilon * np.power(sigma, 6) / np.power(r, 7)
    
r = np.linspace(3.5, 8, 100)

fig, ax = plt.subplots()

ax.plot(r, ljp(r, 0.0103, 3.3))

ax.grid(True, which='both')
ax.axhline(y=0, color='k')

ax.set_xlabel('distance')
ax.set_ylabel('energy')
```

```python
#@markdown 部分的に連続な関数の例
# ref https://stackoverflow.com/a/42838056/13530133
import numpy as np
import matplotlib.pyplot as plt

# Choose evenly spaced x intervals
x = np.arange(-2*np.pi, 2*np.pi, 0.1)

# Compute the tangent for each point
y = np.tan(x)

# Insert a NaN where the difference between successive points is negative
y[:-1][np.diff(y) < 0] = np.nan

# Plot the resulting discontinuous line
plt.plot(x, y)

# Set the range of the axes
plt.axis([-2*np.pi, 2*np.pi, -2, 2])

# Include a title
plt.title('y = tan(x)')

# Optional grid-lines
plt.grid()
```

### 1.2.1 物理量と微分

\begin{align}
\Delta x \to 0; \quad \Delta f \to 0
\end{align}
If $|\Delta x| << 1$, we can write
\begin{align}
\Delta f 
&= f(x+\Delta x) - f(x) = c(x, \Delta x) \Delta x \\
&\sim y = a \ x
\end{align}

```python
#@title 局所的な傾きとしての微分
%%svg
https://upload.wikimedia.org/wikipedia/commons/c/c1/Wiki_slope_in_2d.svg
```

\begin{align}
\lim_{\Delta x \to 0} c (x,\Delta x) 
&= \lim_{\Delta x \to 0 } \dfrac{f(x+\Delta x) - f(x)}{\Delta x}
= \lim_{\Delta x \to 0 } \dfrac{\Delta f}{\Delta x}
\leftarrow \partial _{+}f(x) = \partial _{-}f(x) 
\\
&= \dfrac{df}{dx} := f'(x)
\end{align}

#### 引用
> 物理量を表す関数 $f(x)$ は, 特別な点を除けば, その変数に関して微分可能であり, その導関数 $f'(x)$ は連続関数である

##### [滑らかな関数](https://ja.wikipedia.org/wiki/%E6%BB%91%E3%82%89%E3%81%8B%E3%81%AA%E9%96%A2%E6%95%B0)

- $C^1$: $f'(x)$, $f(x)$は, 1回微分しても滑らなか
- $C^n$: $f^{(n)}(x)$, $f(x)$は, $n$ 回微分しても滑らか
- $C^{\infty}$: $f^{(\infty)}(x)$, $f(x)$は, $\infty$ 回微分しても滑らか

#### [ランダウの記号](https://ja.wikipedia.org/wiki/%E3%83%A9%E3%83%B3%E3%83%80%E3%82%A6%E3%81%AE%E8%A8%98%E5%8F%B7)/[Big O notation](https://en.wikipedia.org/wiki/Big_O_notation)

at round $|\Delta x| <<1$,
\begin{align}
f'(x) 
&\approx \dfrac{f(x+\Delta x) - f(x)}{\Delta x}\\[1mm]
%
f(x+\Delta x) 
&\approx f(x) + f'(x) \Delta x\\[3mm]
%
f(x+\Delta x) 
&= f(x) + f'(x) \Delta x + O\left( \left( \Delta x\right)^2 \right)
\end{align}

### 1.2.2 Taylor 級数

at round $|\Delta x| <<1$,
\begin{align}
f(x+\Delta x) 
&= f(x) + f'(x)\ \Delta x + O\left( \left( \Delta x\right)^2 \right)\\[2mm]
&= f(x) + f'(x)\  \Delta x 
+ d(x)\ \left( \Delta x\right)^2 + O\left( \left( \Delta x\right)^3 \right)\\
\end{align}


\begin{align}
f(x + 1 \Delta x) \approx f(x + 2 \Delta x)
&= f(x + \Delta x + \Delta x)\\[2mm]
f\left(\left(x + \Delta x\right) + \Delta x\right)
&=
f(x+\Delta x) + f'(x+\Delta)\  \Delta x 
+ d(x+\Delta x)\ \left( \Delta x\right)^2 + O\left( \left( \Delta x\right)^3 \right)\\[1mm]
%
&\begin{cases}
f(x+\Delta x) 
&= f(x) + f'(x)\  \Delta x 
+ d(x)\ \left( \Delta x\right)^2 + O\left( \left( \Delta x\right)^3 \right)\\
%
f'(x+\Delta x) 
&=f'(x) + f''(x) \Delta x + O\left( \left( \Delta x\right)^2 \right)\\
d(x+\Delta x) 
&=d(x) + d'(x) \Delta x + O\left( \left( \Delta x\right)^2 \right)
\end{cases}
\end{align}

$\require{cancel}$
\begin{align}
f(x + 2 \Delta x)
=&
\left(
  f(x) + f'(x)\  \Delta x 
+ d(x)\ \left( \Delta x\right)^2 + O\left( \left( \Delta x\right)^3 \right)
\right)\\
&+ \left(
   f'(x) + f''(x) \Delta x + O\left( \left( \Delta x\right)^2 \right)
   \right) \Delta x\\
&+ \left(
  d(x) + d'(x) \Delta x + O\left( \left( \Delta x\right)^2 \right)
  \right)  \left( \Delta x\right)^2\\[1mm]
=&
f(x) 
+ f'(x)\ (2\Delta x )^1
+ \frac{1}{4} \left\{ 2d(x) + f''(x) \right\} \left( 2\Delta x\right)^2
+ O\left( \left( \Delta x\right)^3 \right)\\[2mm]
%
f\left(x + \left( 2 \Delta x\right) \right)
=&
f(x) + f'(x)\  (2\Delta x )^1
+ d(x)\ \left( 2\Delta x\right)^2 + O\left( \left( \bcancel{2}\Delta x\right)^3 \right)
\end{align}

\begin{align}
\left\{ 2d(x) + f''(x) \right\} = d(x) \to d(x) = \dfrac{1}{2}f''(x)
\end{align}

\begin{align}
f(x+\Delta x) 
&= f(x) + f'(x)\  \Delta x 
+ \dfrac{1}{2}f''(x)\ \left( \Delta x\right)^2 + O\left( \left( \Delta x\right)^3 \right)\\
\end{align}

$f(x)$ is $C^{\infty}$,
\begin{align}
f(x+\Delta x) 
&= f(x) + f'(x)\  \Delta x 
+ \dfrac{1}{2!}f''(x)\ \left( \Delta x\right)^2
+ \dfrac{1}{3!}f^{(3)}(x)\ \left( \Delta x\right)^3
+ \cdots
+ \dfrac{1}{n!}f^{(n)}(x)\ \left( \Delta x\right)^n
+ \cdots.
\end{align}

Since RHS is a infinite series, $\Delta x$ has limataion so as to converge the RHS.

\begin{align}
&\begin{cases}
x &\to a\\
\Delta x &\to x - a
\end{cases}\\
%
f\left(a+ (x- a) \right)
 = f(x)
&= f(a) + f'(z)\  \left(x - a \right)
+ \dfrac{1}{2!}f''(a)\ \left(x - a \right)^2
+ \dfrac{1}{3!}f^{(3)}(a)\ \left(x - a \right)^3
+ \cdots
+ \dfrac{1}{n!}f^{(n)}(a)\ \left(x - a \right)^n
+ \cdots.
\end{align}
Here is [Taylor seies](https://en.wikipedia.org/wiki/Taylor_series) for $f(x)$ at around $x=a$.
- at $x=0$, it is called as **Maclaurin series**.

```python
#@markdown Taylor series for $e^x$
# ref. https://watlab-blog.com/2020/05/05/sympy-taylor-series/
%matplotlib inline
%config InlineBackend.figure_formats = ['svg']
from sympy import*
import numpy as np
from matplotlib import pyplot as plt

# 理想波形を生成
cal_x = np.arange(-10.0, 10.1, 0.1)
cal_y = np.exp(cal_x)

# シンボルを定義
x = Symbol('x')

# 任意点周りのテイラー展開を精度違いで計算する
sol = []                                                # 結果の入れ物
a = 0                                                   # 任意点
for i in range(6):
    taylor = series(exp(x), x=x, x0=a, n=i+2).removeO() # removeO()で剰余項を除去する
    display(taylor)                                     # 式を表示させて確認
    taylor_y = lambdify(x, taylor, 'numpy')             # numpyの関数に変換
    sol.append(taylor_y(cal_x))                         # 関数に値を入れて曲線を計算

# ここからグラフ描画---------------------------------------------------
# フォントの種類とサイズを設定する。
plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Times New Roman'

# 目盛を内側にする。
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

# グラフの上下左右に目盛線を付ける。
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')

# 軸のラベルとスケールを設定する。
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim(-2, 2)
ax1.set_ylim(-2, 8)

# データをプロットする。
ax1.plot(cal_x, cal_y, label='Theory', lw=3, color='black')
for j in range(len(sol)):
    ax1.plot(cal_x, sol[j], label='Series:n=' + str(j+2))

fig.tight_layout()

# グラフを表示する。
plt.legend()
plt.show()
plt.close()
# ---------------------------------------------------------------------
```

\begin{array}{l}
f(x)
&= f\left(x_{0}\right)+\lim _{N \to \infty}\left\{f^{\prime}\left(x_{0}\right) \Delta x+f^{\prime}\left(x_{0}+\Delta x\right) \Delta x+\cdots+f^{\prime}\left(x_{0}+(N-1) \Delta x\right) \Delta x\right\},\\[1mm]

## 1.3 多変数関数の微分と積分

### 1.3.1 偏微分

https://en.wikipedia.org/wiki/Partial_derivative

偏微分と偏導関数の定義確認
\begin{align}
\frac {\partial f}{\partial x}
:=&
\lim _{\Delta x\to 0}{\frac {f(x+\Delta x,y)-f(x,y)}{\Delta x}}
=\partial _{x}f(x,y)=f_{x}(x,y)=f_{x}\newline
%
{\frac {\partial f}{\partial y}}
:=&\lim _{\Delta y\to 0}{\frac {f(x,y+\Delta y)-f(x,y)}{\Delta y}}
=\partial _{y}f(x,y)=f_{y}(x,y)=f_{y}
\end{align}

$f(x,y)$についての高次微分に関しても
\begin{array}{ll}
\frac{\partial^{2} f}{\partial x^{2}}=f_{x x}(x, y), & \frac{\partial^{2} f}{\partial y^{2}}=f_{y y}(x, y) \\
\frac{\partial^{2} f}{\partial x \partial y}=f_{x y}(x, y), & \frac{\partial^{2} f}{\partial y \partial x}=f_{y x}(x, y).
\end{array}

$f_{x y}(x, y)$ および $f_{y x}(x, y)$ が変数 $x,y$ について連続ならば
\begin{align}
f_{x y}(x, y) = f_{y x}(x, y)
\end{align}

#### 全微分

\begin{aligned}
\Delta f &=f(x+\Delta x, y+\Delta y)-f(x, y) \\
&=\{f(x+\Delta x, y+\Delta y)-f(x, y+\Delta y)\}+\{f(x, y+\Delta y)-f(x, y)\}\\[2mm]
%
&=f_{x}(x, y+\Delta y) \Delta x+f_{y}(x, y) \Delta y
+ O((\Delta x)^2,(\Delta y)^2)\\[2mm]
\end{aligned}

$\Delta y \to 0, \Delta x \to 0$を順次取れば, 二次の微少量は影響を失い
\begin{aligned}
d f &= \frac {\partial f}{\partial x} d x + \frac {\partial f}{\partial y} d y
+ \cancelto{0}{O((\Delta x)^2,(\Delta y)^2)}\\[2mm]
 &= f_x d x + f_y d y 
\end{aligned}

#### $d^2f$

\begin{aligned}
d^{2} f &=d(d f)=\left(d f_{x}\right) d x+\left(d f_{y}\right) d y \\
&=f_{x x}(x, y) \left( d x\right)^{2}+2 f_{x y}(x, y) d x d y+f_{y y}(x, y) \left(d y\right)^{2}
\end{aligned}

##### $f(x+d x ,y+d y)$ の展開

\begin{align}
f(x+d x ,y+d y) = f(x,y) + df + \frac{1}{2!}d^2f + \frac{1}{3!}d^3f + \cdots 
= \sum_{n=0} \dfrac{1}{n!} d^{(n)}f
\end{align}
